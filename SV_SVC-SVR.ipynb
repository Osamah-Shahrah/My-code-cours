```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import pandas as pd
```

```python
my_data=np.array([[3,2],[6,6],[2.6,3],[7,8],[3.5,5],[6,11]])
my_predict=[0,1,0,1,0,1]


#my_data2=pd.read_csv('house_practice.csv')
#my_data=np.array(my_data2['Size']).reshape(-1, 1)
#my_predict=np.array(my_data2['Price'])



my_model=svm.SVC(kernel='linear',C=1.0)
my_model.fit(my_data,my_predict)
```

```python
print("The predict for thes data [0.5,0.8]  is :",my_model.predict([[0.5,0.8]]))
print("The predict for this data [8.5,10] is :",my_model.predict([[8.5,10]]))

plt.scatter(my_data[:,0],my_data[:,1],c=my_predict)
plt.scatter(0.5,0.8,c ="r")
plt.scatter(8.5,10,c='r')
plt.show
```

```python
#--------------SVR--------------
```

```python
my_model2=svm.SVR(kernel='linear',C=1.0)
my_model2.fit(my_data,my_predict)
pre=my_model2.predict([[0.5,.8]])
pre2=my_model2.predict([[8.5,10]])
```

```python
print("The data predict from the this [0.5,.8] data is:",pre)
print("The data predict from the this [8.5,10] data is:",pre2)

plt.scatter(my_data[:,0],my_data[:,1])
plt.scatter(0.5,0.8,c='r')
plt.scatter(8.5,10,c='r')
plt.show
```
