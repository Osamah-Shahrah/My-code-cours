```python
#------------------Cluster-KMeans----------------------
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import style
style.use("ggplot")
```

```python
my_data=np.array([[3,2],[6,6],[2.6,3],[7,8],[3.5,5],[6,11]])
my_model=KMeans(n_clusters=2)
my_model.fit(my_data)
```

```python
print("The cluster center is",my_model.cluster_centers_)
print("The leabel is :",my_model.labels_)

plt.scatter(my_data[:,0],my_data[:,1],c=my_model.labels_)
plt.scatter(my_model.cluster_centers_[:,0],my_model.cluster_centers_[:,1],marker="X",c='r',s=250,linewidths=5)
plt.show()
```

```python
#-------------------------Dimension reduction---PCA  -----------
```

```python
from sklearn.decomposition import PCA
```

```python
rn=np.random.RandomState(1)
my_data3=np.dot(rn.rand(2,2),rn.randn(2,100)).T

my_model3=PCA(n_components=1)
my_model3.fit(my_data3)

my_data_PCA=my_model3.transform(my_data3)
```

```python
print("The Orgenal shape :",my_data3.shape)
print("The transformed shape :",my_data_PCA.shape)

my_data_newvalues=my_model3.inverse_transform(my_data_PCA)
plt.scatter(my_data3[:,0],my_data3[:,1])
plt.scatter(my_data_newvalues[:,0],my_data_newvalues[:,1])
```
